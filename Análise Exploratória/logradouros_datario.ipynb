```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('logradouros_datario.csv')
df
```

O dataset é grande: 132.064 trechos de logradouros e 23 colunas, Cada linha representa um trecho de via, não um logradouro completo, por isso uma mesma rua pode aparecer várias vezes com trechos diferentes.

Os valores nulos são numerosos em várias colunas e merecem atenção. A coluna nobreza tem mais de 115 mil nulos (87% do total), o que indica que o atributo de nobreza de endereço simplesmente não existe para a maioria dos trechos — não é um dado faltante por erro, mas por inaplicabilidade. As colunas de numeração predial (np_ini_par, np_fin_par, np_ini_imp, np_fin_imp) também têm muitos nulos, pois nem todo trecho possui numeração registrada — como praças, viadutos e túneis.

A variável mais relevante para análise é velocidade_regulamentada, que é numérica contínua e não tem nulos significativos. As variáveis categóricas mais interessantes são tipo_logra_ext, hierarquia e tipo_trecho.

###Analise Univariada

```python
# Box plot da velocidade regulamentada
df[['velocidade_regulamentada']].plot(kind='box', figsize=(6, 5))
plt.title('Box Plot — Velocidade Regulamentada (km/h)', fontsize=13)
plt.ylabel('Velocidade (km/h)')
plt.tight_layout()
plt.show()
```

```python
# Histograma da velocidade regulamentada
df['velocidade_regulamentada'].hist(figsize=(9, 5), bins=20, color='steelblue', edgecolor='white')
plt.title('Histograma — Velocidade Regulamentada (km/h)', fontsize=13)
plt.xlabel('Velocidade (km/h)')
plt.ylabel('Frequência (nº de trechos)')
plt.tight_layout()
plt.show()
```

```python
# Contagem por tipo de logradouro
print('Tipos de logradouro:')
print(df['tipo_logra_ext'].value_counts().head(15))
```

```python
# Gráfico — top 10 tipos de logradouro
df['tipo_logra_ext'].value_counts().head(10).plot(kind='bar', figsize=(12, 5), color='steelblue', edgecolor='white')
plt.title('Top 10 Tipos de Logradouro por Número de Trechos', fontsize=13)
plt.xlabel('Tipo de Logradouro')
plt.ylabel('Número de Trechos')
plt.xticks(rotation=30, ha='right')
plt.tight_layout()
plt.show()
```

```python
# Contagem por hierarquia viária
print('Hierarquia viária:')
print(df['hierarquia'].value_counts())
```

```python
# Gráfico — hierarquia viária
df['hierarquia'].value_counts().plot(kind='bar', figsize=(10, 5), color='coral', edgecolor='white')
plt.title('Trechos por Hierarquia Viária', fontsize=13)
plt.xlabel('Hierarquia')
plt.ylabel('Número de Trechos')
plt.xticks(rotation=20, ha='right')
plt.tight_layout()
plt.show()
```

```python
# Top 10 bairros com mais trechos
print('Top 10 bairros com mais trechos de logradouro:')
print(df['bairro'].value_counts().head(10))
```

```python
# Gráfico — top 10 bairros
df['bairro'].value_counts().head(10).plot(kind='bar', figsize=(12, 5), color='mediumseagreen', edgecolor='white')
plt.title('Top 10 Bairros com Mais Trechos de Logradouro', fontsize=13)
plt.xlabel('Bairro')
plt.ylabel('Número de Trechos')
plt.xticks(rotation=30, ha='right')
plt.tight_layout()
plt.show()
```

```python
# Distribuição de velocidades regulamentadas
print('Distribuição de velocidades regulamentadas:')
print(df['velocidade_regulamentada'].value_counts().sort_index())
```

As Ruas dominam o dataset de forma esmagadora: são mais de 91 mil trechos, enquanto Avenidas somam apenas 13 mil, o que reflete bem a malha viária do Rio, composta majoritariamente por vias locais e residenciais.

Na hierarquia viária, a maioria dos trechos é classificada como Local (82 mil trechos), ou seja, vias de menor fluxo e acesso a imóveis. Vias estruturais e arteriais primárias somam menos de 6 mil trechos, são poucas, mas são as que carregam o maior volume de tráfego da cidade.

Os bairros com mais trechos são Campo Grande, Santa Cruz e Guaratiba, todos no Oeste da cidade, que é a região mais extensa territorialmente. Esse resultado é consistente com o que já vimos no dataset de bairros.

No histograma de velocidade, fica claro que 30 km/h é a velocidade mais comum (mais de 75 mil trechos), seguida de 0 km/h (trechos sem velocidade definida, como praças e áreas não endereçáveis) e 20 km/h. Velocidades acima de 60 km/h são raras e concentradas em vias estruturais.

###Analise Multivariada

```python
# Velocidade média por hierarquia
vel_hierarquia = df.groupby('hierarquia')['velocidade_regulamentada'].mean().sort_values(ascending=False)
vel_hierarquia.plot(kind='bar', figsize=(10, 5), color='steelblue', edgecolor='white')
plt.title('Velocidade Média Regulamentada por Hierarquia Viária', fontsize=13)
plt.xlabel('Hierarquia')
plt.ylabel('Velocidade Média (km/h)')
plt.xticks(rotation=20, ha='right')
plt.tight_layout()
plt.show()
```

```python
# Velocidade média por tipo de logradouro (top 10)
vel_tipo = df.groupby('tipo_logra_ext')['velocidade_regulamentada'].mean().sort_values(ascending=False).head(10)
vel_tipo.plot(kind='bar', figsize=(12, 5), color='darkorange', edgecolor='white')
plt.title('Velocidade Média por Tipo de Logradouro (Top 10)', fontsize=13)
plt.xlabel('Tipo de Logradouro')
plt.ylabel('Velocidade Média (km/h)')
plt.xticks(rotation=30, ha='right')
plt.tight_layout()
plt.show()
```

```python
# Box plot — velocidade por hierarquia
plt.figure(figsize=(12, 6))
sns.boxplot(x='hierarquia', y='velocidade_regulamentada', data=df, palette='pastel',
            order=['Local', 'Coletora', 'Arterial secundária', 'Arterial primária', 'Estrutural'])
plt.title('Distribuição de Velocidade Regulamentada por Hierarquia Viária', fontsize=13)
plt.xlabel('Hierarquia')
plt.ylabel('Velocidade (km/h)')
plt.xticks(rotation=20, ha='right')
plt.tight_layout()
plt.show()
```

```python
# Scatter plot — numeração predial inicial vs final (par)
df_num = df[['np_ini_par', 'np_fin_par']].dropna()
df_num = df_num[df_num['np_fin_par'] < 5000]  # remove outliers extremos para visualização

sns.scatterplot(x='np_ini_par', y='np_fin_par', data=df_num, alpha=0.3)
plt.title('Numeração Predial Inicial vs Final (lado par)', fontsize=13)
plt.xlabel('Nº Inicial (par)')
plt.ylabel('Nº Final (par)')
plt.tight_layout()
plt.show()
```

```python
# Pair plot das variáveis numéricas relevantes
cols_pair = ['velocidade_regulamentada', 'np_ini_par', 'np_fin_par']
sns.pairplot(df[cols_pair].dropna())
plt.suptitle('Pair Plot — Variáveis Numéricas', y=1.02, fontsize=13)
plt.tight_layout()
plt.show()
```

O gráfico de velocidade média por hierarquia viária confirma exatamente o que se espera do planejamento urbano: existe uma escada de velocidades muito bem definida, vias estruturais chegam a quase 70 km/h de média, enquanto vias locais ficam em torno de 24 km/h. Isso mostra que a classificação hierárquica do dataset é consistente com a realidade do trânsito.

No box plot, chama atenção a alta dispersão nas vias estruturais, há trechos estruturais com velocidade baixa (cruzamentos, regiões de transição) e outros com velocidade muito alta, o que gera uma caixa mais larga. Vias locais têm dispersão menor, concentradas entre 20 e 30 km/h.

No scatter plot da numeração predial, os pontos seguem uma linha diagonal clara: o número final é sempre maior que o inicial, formando um padrão linear previsível. Isso é matematicamente esperado mas confirma a consistência do dado. Os trechos mais longos aparecem no canto superior direito, com numerações que chegam a 5.000.

O pair plot é rico aqui, diferente dos datasets anteriores, temos três variáveis numéricas contínuas com relações interpretáveis entre si.

###Correlação e Heatmap

```python
# Matriz de correlação
cols_corr = ['velocidade_regulamentada', 'np_ini_par', 'np_fin_par']
df_corr = df[cols_corr].dropna()
df_corr.corr()
```

```python
# Heatmap de correlação
plt.figure(figsize=(7, 5))
sns.heatmap(df_corr.corr(), annot=True, fmt='.2f', cmap='coolwarm', center=0)
plt.title('Heatmap de Correlação — Variáveis Numéricas', fontsize=13)
plt.tight_layout()
plt.show()
```

velocidade_regulamentada vs numeração predial apresenta correlação fraca, indicando que a velocidade de uma via tem pouca relação com onde ela começa ou termina numericamente. A velocidade é determinada pela hierarquia e pelo tipo da via, não pela localização do trecho na sequência de numeração.

###Conclusões

1.   Os nulos são abundantes mas justificados: nobreza é inaplicável para a maioria dos trechos, e as colunas de numeração predial não existem para praças, viadutos e áreas não endereçáveis.
2.   Ruas são o tipo de logradouro dominante (91 mil trechos), e a hierarquia Local representa mais de 82 mil trechos, a cidade é feita de vias pequenas.
3. A velocidade mais comum é 30 km/h, e existe uma escada de velocidades perfeitamente consistente com a hierarquia viária: quanto mais importante a via, maior a velocidade regulamentada.
4. Campo Grande, Santa Cruz e Guaratiba lideram em número de trechos, reforçando que o Oeste da cidade é a região mais extensa e com maior malha viária.
5. A correlação entre numeração predial inicial e final é fortíssima (0.96), enquanto a velocidade tem baixa correlação com a numeração, confirmando que são dimensões independentes da via.
